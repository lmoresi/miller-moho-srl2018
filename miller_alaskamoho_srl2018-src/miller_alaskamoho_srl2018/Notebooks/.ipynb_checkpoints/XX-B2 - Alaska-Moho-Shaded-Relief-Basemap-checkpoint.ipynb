## Basemap for overlaying moho plots

We use a shaded relief image from the Natural Earth Website for this. It's the same resolution as ETOPO1 but is supposedly
slightly cleaner. 

```python
import os.path as path
import stripy as stripy
import numpy as np

import cartopy
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
```

```python
## Assuming the earlier notebook was run ...

data = np.load("AlaskaMoho.npz")
moho_data = data['alaska_moho']
```

```python
## Grid the original data
```

```python
datagrid = stripy.sTriangulation(np.radians(moho_data['lon']), np.radians(moho_data['lat']), permute=True)
print("Points in triangulation: {}".format(datagrid.npoints))
```

```python
## Make a background image to merge with these maps (SLOW !!)

import gdal

globalsrelief      = gdal.Open("ShadedRelief/GRAY_HR_SR_OB.tif")
globalsrelief_img  = globalsrelief.ReadAsArray()  # .transpose(1,2,0)

global_extent = [-180,180,-90,90]


fig = plt.figure(figsize=(15, 8), facecolor=None)
ax1  = plt.subplot(111, projection=ccrs.AlbersEqualArea(central_longitude=-154, central_latitude=50, 
                                                       standard_parallels=(55,65) ))
grat = cartopy.feature.NaturalEarthFeature(category="physical", scale="10m",
                                name="graticules_5")

ax1.add_feature(grat, linewidth=0.5, 
                  linestyle="--",
                  edgecolor="#88888888", 
                  facecolor="None", zorder=2)



ax1.set_extent([-175,-130,50,72])

lons = np.degrees(datagrid.lons)
lats = np.degrees(datagrid.lats)

m=ax1.tricontourf(lons, lats, datagrid.simplices, lons, 2,
               cmap=plt.cm.RdYlBu, alpha=0.0, vmin=25, vmax=50, 
               transform=ccrs.PlateCarree(), zorder=0)

## Frustrating, but we need a colorbar or the images are not aligned 

cb1 = plt.colorbar(ax=ax1, mappable=m, shrink=0.5, extend='both', drawedges=False, alpha=0.0)
cb1.set_ticks([0.0], update_ticks=True)
cb1.outline.set_color("White")

ax1.imshow(globalsrelief_img, origin='upper', transform=ccrs.PlateCarree(),
          extent=global_extent, zorder=1, alpha=1.0, cmap=plt.cm.Greys_r, vmin=0.0, vmax=255)



grat = cartopy.feature.NaturalEarthFeature(category="physical", scale="10m",
                                    name="graticules_5")

ax1.set_extent([-175,-130,50,72])
ax1.coastlines(resolution="10m",color="#000000", linewidth=0.25, zorder=99)
ax1.add_feature(grat, linewidth=1.0, 
                      linestyle="--",
                      edgecolor="#88888888", 
                      facecolor="None", zorder=2)

fig.savefig("AlaskaMohoImgBg.png", dpi=600)
```
